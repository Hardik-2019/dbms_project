```python
#import libraries
from tkinter import *
import tkinter.ttk as ttk
from PIL import ImageTk, Image
import tkinter.messagebox as tkMessageBox
import pandas as pd
import sqlite3

#function to define database
def Database():
    global conn, cursor
    #creating student database
    conn = sqlite3.connect("student.db")
    cursor = conn.cursor()
    #creating STUD_REGISTRATION table
    cursor.execute(
        "CREATE TABLE IF NOT EXISTS STUD_REGISTRATION (STU_ID INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, STU_NAME TEXT, STU_CONTACT TEXT, STU_EMAIL TEXT, STU_ROLLNO TEXT, STU_BRANCH TEXT)")
    cursor.execute(
        "CREATE TABLE IF NOT EXISTS Subjects (Subject_code varchar(100) primary key, Subject_name varchar(100), Subject_credit int)")
    cursor.execute(
        "CREATE TABLE IF NOT EXISTS ACAD_DETAILS (SERIAL_NO INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,STU_ID INTEGER, SUB_CODE TEXT, SEM INTEGER)")
```

```python
#function to insert data into database
def register():
    Database()
    #getting form data
    name1=name.get()
    con1=contact.get()
    email1=email.get()
    rol1=rollno.get()
    branch1=branch.get()
    #applying empty validation
    if name1=='' or con1==''or email1=='' or rol1==''or branch1=='':
        tkMessageBox.showinfo("Warning","fill the empty field!!!")
    else:
        #execute query
        conn.execute('INSERT INTO STUD_REGISTRATION (STU_NAME,STU_CONTACT,STU_EMAIL,STU_ROLLNO,STU_BRANCH) \
              VALUES (?,?,?,?,?)',(name1,con1,email1,rol1,branch1));
        conn.commit()
        tkMessageBox.showinfo("Message","Stored successfully")
        #refresh table data
        DisplayData()
        conn.close()

def Reset():
    #clear current data from table
    tree.delete(*tree.get_children())
    #refresh table data
    DisplayData()
    #clear search text
    SEARCH.set("")
    name.set("")
    contact.set("")
    email.set("")
    rollno.set("")
    branch.set("")

    
def updateCol():
    name1=name.get()
    con1=contact.get()
    email1=email.get()
    rol1=rollno.get()
    branch1=branch.get()

#applying empty validation
    if name1=='' or con1==''or email1=='' or rol1==''or branch1=='':
        tkMessageBox.showinfo("Warning","fill the empty field!!!")
    else:
        Database()
        curItem = tree.focus()
        contents = (tree.item(curItem))
        selecteditem = contents['values']
        id1 = selecteditem[0] 
        params = (name1,con1,email1,rol1,branch1,id1)
        cursor=conn.execute("UPDATE STUD_REGISTRATION SET STU_NAME=?, STU_CONTACT=?, STU_EMAIL=?, STU_ROLLNO=?, STU_BRANCH=? WHERE STU_ID=?",params)       
        #cursor=conn.execute("UPDATE STUD_REGISTRATION SET STU_NAME=%s,STU_CONTACT=%s,STU_EMAIL=%s,STU_ROLLNO=%s,STU_BRANCH=%s WHERE STU_ID = %d" % (name1,con1,email1,rol1,branch1,selecteditem[0]))
        #cursor=conn.execute("UPDATE STUD_REGISTRATION SET name1=STU_NAME,STU_CONTACT=con1,STU_EMAIL=email1,STU_ROLLNO=rol1,STU_BRANCH=branch1 WHERE STU_ID = %d" % selecteditem[0])
        #id = selecteditem[0]
        #params = (name1,con1,email1,rol1,branch1,id)
        #cursor=conn.execute("UPDATE STUD_REGISTRATION SET VALUES (?, ?, ?, ?, ?) WHERE STU_ID (?)", params)
        conn.commit()
        DisplayData()
        conn.close()
        cursor.close()

def Update():
    #open database
    Database()
    if not tree.selection():
        tkMessageBox.showwarning("Warning","Select data to update")
    else:
        result = tkMessageBox.askquestion('Confirm', 'Are you sure you want to update    this record?',
                                          icon="warning")
        if result == 'yes':
            curItem = tree.focus()
            contents = (tree.item(curItem))
            selecteditem = contents['values']
            #tree.delete(curItem)
            cursor=conn.execute("SELECT * FROM STUD_REGISTRATION WHERE STU_ID = %d" % selecteditem[0])
            SEARCH.set("")
            name.set(selecteditem[1])
            contact.set(selecteditem[2])
            email.set(selecteditem[3])
            rollno.set(selecteditem[4])
            branch.set(selecteditem[5])

            conn.commit()
            cursor.close()
            conn.close()

def Delete():
    #open database
    Database()
    if not tree.selection():
        tkMessageBox.showwarning("Warning","Select data to delete")
    else:
        result = tkMessageBox.askquestion('Confirm', 'Are you sure you want to delete this record?',
                                          icon="warning")
        if result == 'yes':
            curItem = tree.focus()
            contents = (tree.item(curItem))
            selecteditem = contents['values']
            tree.delete(curItem)
            cursor=conn.execute("DELETE FROM STUD_REGISTRATION WHERE STU_ID = %d" % selecteditem[0])
            conn.commit()
            cursor.close()
            conn.close()

#function to search data
def SearchRecord():
    #open database
    Database()
    #checking search text is empty or not
    if SEARCH.get() != "":
        #clearing current display data
        tree.delete(*tree.get_children())
        #select query with where clause
        cursor=conn.execute("SELECT * FROM STUD_REGISTRATION WHERE STU_NAME LIKE ?", ('%' + str(SEARCH.get()) + '%',))
        #fetch all matching records
        fetch = cursor.fetchall()
        #loop for displaying all records into GUI
        for data in fetch:
            tree.insert('', 'end', values=(data))
        cursor.close()
        conn.close()

#defining function to access data from SQLite database
def DisplayData():
    #open database
    Database()
    #clear current data
    tree.delete(*tree.get_children())
    #select query
    cursor=conn.execute("SELECT * FROM STUD_REGISTRATION")
    #fetch all data from database
    fetch = cursor.fetchall()
    #loop for displaying all data in GUI
    for data in fetch:
        tree.insert('', 'end', values=(data))
    cursor.close()
    conn.close()

```

```python
#function to insert data into database
def register2():
    Database()
    #getting form data
    
    st_id1 = st_id.get()
    sub_code1=sub_code.get()
    sem1=sem.get()

    #applying empty validation
    if st_id1=='' or sub_code1==''or sem1=='':
        tkMessageBox.showinfo("Warning","fill the empty field!!!")
    else:
        #execute query
        conn.execute('INSERT INTO ACAD_DETAILS (STU_ID, SUB_CODE, SEM) \
              VALUES (?,?,?)',(st_id1,sub_code1,sem1));
        conn.commit()
        tkMessageBox.showinfo("Message","Stored successfully")
        #refresh table data
        DisplayData2()
        conn.close()

def Reset2():
    #clear current data from table
    tree2.delete(*tree2.get_children())
    #refresh table data
    DisplayData2()
    #clear search text
    SEARCH2.set("")
    st_id.set("")
    sub_code.set("")
    sem.set("")

    
def updateCol2():
    st_id1=st_id.get()
    sub_code1=sub_code.get()
    sem1=sem.get()

#applying empty validation
    if st_id1=='' or sub_code1==''or sem1=='':
        tkMessageBox.showinfo("Warning","fill the empty field!!!")
    else:
        Database()
        curItem = tree2.focus()
        contents = (tree2.item(curItem))
        selecteditem = contents['values']
        id1 = selecteditem[0] 
        params = (st_id1,sub_code1,sem1,id1)
        cursor=conn.execute("UPDATE ACAD_DETAILS SET STU_ID=?, SUB_CODE=?, SEM=? WHERE SERIAL_NO=?",params)       
        #cursor=conn.execute("UPDATE STUD_REGISTRATION SET STU_NAME=%s,STU_CONTACT=%s,STU_EMAIL=%s,STU_ROLLNO=%s,STU_BRANCH=%s WHERE STU_ID = %d" % (name1,con1,email1,rol1,branch1,selecteditem[0]))
        #cursor=conn.execute("UPDATE STUD_REGISTRATION SET name1=STU_NAME,STU_CONTACT=con1,STU_EMAIL=email1,STU_ROLLNO=rol1,STU_BRANCH=branch1 WHERE STU_ID = %d" % selecteditem[0])
        #id = selecteditem[0]
        #params = (name1,con1,email1,rol1,branch1,id)
        #cursor=conn.execute("UPDATE STUD_REGISTRATION SET VALUES (?, ?, ?, ?, ?) WHERE STU_ID (?)", params)
        conn.commit()
        DisplayData2()
        conn.close()
        cursor.close()

def Update2():
    #open database
    Database()
    if not tree2.selection():
        tkMessageBox.showwarning("Warning","Select data to update")
    else:
        result = tkMessageBox.askquestion('Confirm', 'Are you sure you want to update    this record?',
                                          icon="warning")
        if result == 'yes':
            curItem = tree2.focus()
            contents = (tree2.item(curItem))
            selecteditem = contents['values']
            #tree.delete(curItem)
            cursor=conn.execute("SELECT * FROM ACAD_DETAILS WHERE SERIAL_NO = %d" % selecteditem[0])
            SEARCH2.set("")
            st_id.set(selecteditem[1])
            sub_code.set(selecteditem[2])
            sem.set(selecteditem[3])

            conn.commit()
            cursor.close()
            conn.close()

def Delete2():
    #open database
    Database()
    if not tree2.selection():
        tkMessageBox.showwarning("Warning","Select data to delete")
    else:
        result = tkMessageBox.askquestion('Confirm', 'Are you sure you want to delete this record?',
                                          icon="warning")
        if result == 'yes':
            curItem = tree2.focus()
            contents = (tree2.item(curItem))
            selecteditem = contents['values']
            tree2.delete(curItem)
            cursor=conn.execute("DELETE FROM ACAD_DETAILS WHERE SERIAL_NO = %d" % selecteditem[0])
            conn.commit()
            cursor.close()
            conn.close()

#function to search data
def SearchRecord2():
    #open database
    Database()
    #checking search text is empty or not
    if SEARCH2.get() != "":
        #clearing current display data
        tree2.delete(*tree.get_children())
        #select query with where clause
        cursor=conn.execute("SELECT * FROM ACAD_DETAILS WHERE STU_ID LIKE ?", ('%' + str(SEARCH2.get()) + '%',))
        #fetch all matching records
        fetch = cursor.fetchall()
        #loop for displaying all records into GUI
        for data in fetch:
            tree2.insert('', 'end', values=(data))
        cursor.close()
        conn.close()
#defining function to access data from SQLite database
def DisplayData2():
    #open database
    Database()
    #clear current data
    tree2.delete(*tree2.get_children())
    #select query
    cursor=conn.execute("SELECT * FROM ACAD_DETAILS")
    #fetch all data from database
    fetch = cursor.fetchall()
    #loop for displaying all data in GUI
    for data in fetch:
        tree2.insert('', 'end', values=(data))
    cursor.close()
    conn.close()
```
